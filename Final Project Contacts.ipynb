```python
from os import linesep
class Contact:
    
    def __init__(self, ifirstname='', ilastname='', iphonenumber='', iemail=''):
        self.__firstname = ifirstname
        self.__lastname = ilastname
        self.__phonenumber = iphonenumber
        self.__email = iemail
    
    def get_firstname(self):
        return self.__firstname
    
    def get_lastname(self):
        return self.__lastname
   
    def get_phonenumber(self):
        return self.__phonenumber
    
    def get_email(self):
        return self.__email
    
    def set_firstname(self,ifirstname):
        self.__firstname = ifirstname
    
    def set_lastname(self, ilastname):
        self.__lastname = ilastname
    
    def set_phonenumber(self, iphonenumber):
        self.__phonenumber = iphonenumber
        
    def set_email(self, iemail):
        self.__email = iemail
    
    
    
    def __str__(self):
        return self.__firstname + ', ' + self.__lastname + ',' + self.__phonenumber + ',' + self.__email

```

```python
class Store:

    def __init__(self):
        self.__contact_list = []

    def add_contact(self,new_contact):

        self.__contact_list.append(new_contact)

    def get_all_contact(self):

        return self.__contact_list.copy()
    
    
    def search_contact_by_name(self,pattern):
        
        result = []
        
        for contact in self.__contact_list:
            if pattern.lower() in contact.get_firstname().lower():
                
                result.append (contact)
        for contact in self.__contact_list:
            if pattern.lower() in contact.get_lastname().lower():
                
                result.append (contact)

        return result
    
    
    
    def search_contact_by_email(self,pattern):
        
        result = []
        
        for contact in self.__contact_list:
            if pattern.lower() in contact.get_email().lower():
                
                result.append (contact)
    
        return result
    
    
    def search_contact_by_number(self,pattern):
        
        result = []
        
        for contact in self.__contact_list:
            if pattern in contact.get_phonenumber():
                
                result.append (contact)

        return result
    
    def search_specific_contact(self,ifirstname,ilastname,iphonenumber,iemail):
        found_contact = None 
        
        for contact in self.__contact_list:
            if (ifirstname.lower() in contact.get_firstname().lower() and ilastname.lower() in contact.get_lastname().lower()
            and iemail.lower() in contact.get_email().lower() and iphonenumber in contact.get_phonenumber()):
                found_contact = contact
                break
                
        return found_contact
    
    def edit_contact(self, ifirstname, ilastname, iphonenumber, iemail, new_first= '', new_last = '', new_phone = '', new_email = ''): 
        found_contact = self.search_specific_contact(ifirstname, ilastname, iphonenumber, iemail)
        
        if (new_first != ''):
            found_contact.set_firstname(new_first)
        if (new_last != ''):
            found_contact.set_lastname(new_last)
            
        if (new_phone != ''):
            found_contact.set_phonenumber(new_phone)
        if (new_email != ''):
            found_contact.set_email(new_email)
    
    def remove_contact(self, ifirstname, ilastname, iphonenumber, iemail):
        found_contact = self.search_specific_contact(ifirstname, ilastname, iphonenumber, iemail)
        
        self.__contact_list.remove(found_contact)
```

```python
class Application:
    
    def __init__(self):
        self.__store = Store()
        self.lines = []
        contact_file = open ('contact.txt','r',newline=linesep)
               
        for line in contact_file:
            self.lines.append(line.rstrip(linesep))
        contact_file.close()
        for line in self.lines:
            contact = line.split()
            print(contact)
            if contact != []:
                newcontact=Contact(contact[0],contact[1],contact[2],contact[3])
                self.__store.add_contact(newcontact)
            
            
    def view_all_contact(self):
        all_contact = self.__store.get_all_contact()
        for contact in all_contact:
            print(contact)
            
    def search_contact_by_name(self):
        pattern = input('Please enter a name: ')
        search_contact = self.__store.search_contact_by_name(pattern)
        for contact in search_contact:
            print(contact)
    
    def search_contact_by_number(self):
        pattern = input('Please enter a phone number: ')
        search_contact = self.__store.search_contact_by_number(pattern)
        for contact in search_contact:
            print(contact)
        
    def search_contact_by_email(self):
        pattern = input('Please enter an email: ')
        search_contact = self.__store.search_contact_by_email(pattern)
        for contact in search_contact:
            print(contact)
    
    
    
    def add_contact(self):
        ifirstname = input('Please enter the new first name: ')
        ilastname = input('Please enter the new last name: ')
        iphonenumber = input('Please enter the new number: ')
        iemail = input('Please enter new email: ')
        self.__store.add_contact(Contact(ifirstname, ilastname, iphonenumber, iemail))
        print('Item added')
    
   
     
    def edit_contact(self):
        oldname = input('Please enter the firstname of the contact to be edited: ')
        oldlast=input('Please enter the last name of the contact to be edited')
        oldnumber = input('Please enter the number of the item to be edited: ')
        oldemail = input('Please enter the email to be edited')
        newfirst = input('Please enter the new first name: ')
        newlast=input('Please enter new last name: ')
        newphone = input('Please enter the new phone number: ')
        newemail = input('Please enter the new email: ')
        self.__store.edit_contact(oldname,oldlast,oldnumber,oldemail,new_first=newfirst,new_last=newlast,new_phone = newphone,new_email = newemail)
        
        print('Item edited')
        
    def remove_contact(self):
        firstname = input('Please enter the first name of the contact to be removed: ')
        lastname = input('Please enter the last name to be removed: ')
        phonenumber = input('Please enter the number of the contact to be removed: ')
        email = input('Please enter the email of the contact to be removed: ')
        self.__store.remove_contact(firstname,lastname,phonenumber,email)
        print('Item removed')
    
    
    def run (self):
        while(True):
            print()
            print('---------')
            print('Menu ')
            print('---------')
            print('1 - View all Contacts')
            print('2 - Search contacts by Name')
            print('3 - Search contacts by Phone number')
            print('4 - Search contacts by Email')
            print('5 - Add new contact')
            print('6 - Edit a contact')
            print('7 - Remove a contact')
            print('8 - Exit Application')
            
            option = input('Please enter your option: ')
            
            if (option=='1'):
                self.view_all_contact()
            if (option=='2'):
                self.search_contact_by_name()
            if (option=='3'):
                self.search_contact_by_number()
            if (option=='4'):
                self.search_contact_by_email()
            if (option=='5'):
                self.add_contact()
            if (option=='6'):
                self.edit_contact()
            if (option=='7'):
                self.remove_contact()
            if (option =='8'):
              
                array = []
                all_contact = self.__store.get_all_contact()
                
                
                for contact in all_contact:
                    array.append(contact)
                
                contact_file = open ('contact.txt','w')
                print(array)
                
                for contactlist in array:
                    contact_file.write(contactlist.get_firstname()+ ", " +contactlist.get_lastname() + ', ' +contactlist.get_phonenumber() + ", " + contactlist.get_email()+linesep)
                contact_file.close()
                lines = []
          
```

```python
app = Application()
app.run()
```
